```python
import os
import shutil

# Definierte Pfade
path_raw_1 = r'C:\Users\rehan\meine_repos\Masterarbeit\Chapter 3\0_raw_sims\sim_shape_1'
target_path_1 = r'C:\Users\rehan\meine_repos\Masterarbeit\Chapter 3\1_sims\sim_shape_1___B'

# Überprüfen, ob das Quellverzeichnis existiert
if not os.path.exists(path_raw_1):
    raise FileNotFoundError(f"Das Quellverzeichnis '{path_raw_1}' wurde nicht gefunden.")

for B in [500,1000,2000,4000]:
    # Die Zahlen, nach denen gefiltert werden soll
    numbers = [499, 999, 1999, 3999]

    folders_to_copy = []
    # Iteration über alle Ordner im Quellverzeichnis
    for folder in os.listdir(path_raw_1):
        # Überprüfen, ob der Ordnername die Zeichenkette '(B_RF)500' enthält
        if f'(B_RF){B}' in folder:
            source_folder = os.path.join(path_raw_1, folder)
            folders_to_copy.append(source_folder)

    # Kopieren der Ordner in das entsprechende Zielverzeichnis
    for source_folder in folders_to_copy:
        folder_name = os.path.basename(source_folder)

        # Überprüfen, welche Zahl im Ordnernamen enthalten ist
        found_number = None
        for number in numbers:
            if f'n_train{number}' in folder_name:
                found_number = str(number)
                break  # Wenn die Zahl gefunden wurde, Schleife beenden

        if found_number:
            # Zielverzeichnis erstellen (falls nicht vorhanden)
            destination_folder = os.path.join(target_path_1 + str(B), found_number, folder_name)
            os.makedirs(os.path.dirname(destination_folder), exist_ok=True)

            # Falls der Zielordner bereits existiert, optional löschen
            if os.path.exists(destination_folder):
                shutil.rmtree(destination_folder)

            # Ordner kopieren
            try:
                shutil.copytree(source_folder, destination_folder)
                print(f"Ordner '{source_folder}' erfolgreich nach '{destination_folder}' kopiert.")
            except Exception as e:
                print(f"Fehler beim Kopieren von '{source_folder}': {e}")
        else:
            print(f"Keine passende Zahl in '{folder_name}' gefunden. Ordner wird nicht kopiert.")
```
